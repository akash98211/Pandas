```python
import pandas as pd
import numpy as np
```

```python
array = np.random.randint(1,20,size = (6,5))
```

```python
df = pd.DataFrame(array,columns=list('abcde'))
df
```

```python
def sqr(n):
    return n**2

df.apply(sqr)
```

```python
df.apply(np.square)
```

```python
import re
def clean_str(s):
    print('S is ',s)
    s = re.sub('[^A-Za-z]', '',s)
    return s

df = pd.DataFrame({"Country" : ["India 1#",' @ UK','$ USA'],
                  'Capitals':['Delhi','london',"washington"],
                  'Date': ['02 Nov 2021','02 Nov 2021','02 Nov 2021']})
df
df['Country'] = df['Country'].apply(clean_str)
df
```
